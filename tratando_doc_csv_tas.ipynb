Reading CSV

```python
file_path ='/content/drive/MyDrive/Temperatura_Precipitacao_analises/limpos/quartile_for_year/'


file = "tas_seasons_for_year_consolidado.csv"
```

```python
with open(f"{file_path}{file}", 'rb') as encoding:
    r_encoding = chardet.detect(encoding.read(100000))
r_encoding
```

```python

with open (f"{file_path}{file}",'r',encoding=r_encoding['encoding']) as arquivo :
  df_tas = pd.read_csv(arquivo)
```

<font color='blue'> DataFrame

```python
df_tas
```

```python
colunas = [c.strip().lower().replace("50%","tas") for c in df_tas.columns]
df_tas.columns = colunas
```

```python
df_tas
```

```python
df_tas_mediana = pd.pivot_table(df_tas, values =['tas'], columns=['iso3'], index='year')
```

```python
df_tas_mediana
```

```python
df_tas_mediana =df_tas_mediana.swaplevel(i=-2,j=-1,axis=1)
```
